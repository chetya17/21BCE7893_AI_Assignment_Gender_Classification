```python
from tensorflow.keras.callbacks import Callback
class PredictionCallback(Callback):
    def __init__(self, X_test, y_test):
        self.X_test = X_test
        self.y_test = y_test

    def on_epoch_end(self, epoch, logs=None):
        # Get predictions on the test data
        predictions = self.model.predict(self.X_test)

        # Convert probabilities to class labels (use np.argmax)
        predicted_labels = np.argmax(predictions, axis=1)

        # Get true labels (convert one-hot encoded labels back to integers)
        true_labels = np.argmax(self.y_test, axis=1)

        # Print or log the predictions, true labels, and inputs
        print(f'Epoch {epoch + 1} Predictions:')
        for i in range(len(predicted_labels)):
            #print(f'True Label: {true_labels[i]}, Predicted Label: {predicted_labels[i]}')
            if(true_labels[i] != predicted_labels[i]):
              print(f'True Label: {true_labels[i]}, Predicted Label: {predicted_labels[i]}')
              print(f'True Exp: {ldict.get(true_labels[i])}, Predicted Label: {ldict.get(predicted_labels[i])}')

#f'Input: {self.X_test[i]},
# Create an instance of the custom callback
prediction_callback = PredictionCallback(X_test, y_test)
```
